### Simulaton

#### 1. Library

```python
import numpy as np
from PIL import Image
from mujoco_env.y_env import SimpleEnv
```

#### 2. Setup Sim Environment

```python
xml_path = './asset/example_scene_y.xml'
PnPEnv = SimpleEnv(xml_path, seed = 0, state_type = 'joint_angle')
```

#### 3. Run Simulation

```python
PnPEnv.env.reset(step=True)
PnPEnv.env.init_viewer()
while PnPEnv.env.is_viewer_alive():
    PnPEnv.env.step()
    PnPEnv.env.render()
print ("Done.")
```
